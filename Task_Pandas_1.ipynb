# Removing the '$' from item_price

```python
l = []
for i in df['item_price']:
    l.append(i.split('$'))
```

```python
df['item_price'] = [i[1] for i in l]
```

```python
df
```

# Replace NaN with item_name

```python
df['choice_description'].fillna(df['item_name'],inplace=True)
```

```python
df
```

#  Calculating Total _Price:

```python
df['Total_Price'] = pd.to_numeric(df['quantity'] ) * pd.to_numeric(df['item_price'])
```

# Which was the most ordered item? and How many items were ordered?

```python
c = df.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head()
```

```python
c = df.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=True)
c.head()
```

# Maximum Price items

```python
max_price = df.iloc[df[df['item_price'] == max(df['item_price'])].index]
max_price.head()
```

# Manimum Price items

```python
min_price = df.iloc[df[df['item_price'] == min(df['item_price'])].index]
min_price.head()
```

# What was the most ordered item in the choice_description column?

```python
c = df.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head()
```

# How much was the revenue for the period in the dataset?

```python
revenue = (df['quantity']* pd.to_numeric(df['item_price'])).sum()
print('Revenue was: $' + str(revenue))
```
